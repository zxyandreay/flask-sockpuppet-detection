## Semantic Analysis-Based Sockpuppet Detection in Wikipedia

This section explores sockpuppet detection through semantic analysis, focusing on understanding and leveraging the meanings and relationships of words within the Wikipedia dataset. The method employs various NLP techniques to analyze the text data, aiming to identify patterns or anomalies indicative of sockpuppet behavior based solely on content semantics.


```python
# Importing necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC  # Importing the SVM classifier
from sklearn.metrics import classification_report, accuracy_score

# Load the pre-split training and testing sets
train_data = pd.read_csv('wikipedia_sockpuppet_dataset_TRAIN.csv')
test_data = pd.read_csv('wikipedia_sockpuppet_dataset_TEST.csv')

# Combine the training and testing text to fit the vectorizer
all_text = pd.concat([train_data['edit_text'], test_data['edit_text']])

# Semantic Analysis using TF-IDF
tfidf = TfidfVectorizer(max_features=1000)
tfidf.fit(all_text)  # Learn vocabulary and idf from training set
X_train = tfidf.transform(train_data['edit_text']).toarray()
X_test = tfidf.transform(test_data['edit_text']).toarray()

# Labels for the dataset
y_train = train_data['is_sockpuppet']
y_test = test_data['is_sockpuppet']

# Model Training using SVM
model = SVC(random_state=42)
model.fit(X_train, y_train)

# Predictions and Evaluation
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Sentiment Analysis-Based Sockpuppet Detection in Wikipedia

In this part of the notebook, sentiment analysis is used to detect sockpuppets on Wikipedia. The approach analyzes the emotional tone and subjective expressions in user contributions, looking for sentiment patterns that are often associated with deceptive or manipulative online behaviors characteristic of sockpuppets.


```python
# Importing necessary tools and libraries
import pandas as pd
import numpy as np
from textblob import TextBlob
from sklearn.svm import SVC  # Importing the SVM classifier
from sklearn.metrics import classification_report, accuracy_score

# Load the pre-split training and testing datasets
train_data = pd.read_csv('wikipedia_sockpuppet_dataset_TRAIN.csv')
test_data = pd.read_csv('wikipedia_sockpuppet_dataset_TEST.csv')

# Sentiment Analysis directly on the original text
train_data['polarity'] = train_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
train_data['subjectivity'] = train_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
test_data['polarity'] = test_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
test_data['subjectivity'] = test_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Preparing feature arrays
X_train = np.column_stack((train_data['polarity'], train_data['subjectivity']))
X_test = np.column_stack((test_data['polarity'], test_data['subjectivity']))

# Extracting labels
y_train = train_data['is_sockpuppet']
y_test = test_data['is_sockpuppet']

# Model Training using SVM
model = SVC(random_state=42)
model.fit(X_train, y_train)

# Predictions and Evaluation
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Integrated Semantic and Sentiment Analysis for Sockpuppet Detection in Wikipedia

This section combines semantic and sentiment analysis to enhance the accuracy of sockpuppet detection in Wikipedia. By integrating both analytical dimensions, the methodology seeks to provide a comprehensive view of the textual data, tapping into both the explicit meaning of the text (semantics) and the underlying emotions or attitudes (sentiment) to more effectively identify sockpuppet accounts.


```python
# Importing necessary tools and libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob
from sklearn.svm import SVC  # Importing the SVM classifier
from sklearn.metrics import classification_report, accuracy_score

# Load the pre-split training and testing datasets
train_data = pd.read_csv('wikipedia_sockpuppet_dataset_TRAIN.csv')
test_data = pd.read_csv('wikipedia_sockpuppet_dataset_TEST.csv')

# Feature extraction with TF-IDF directly on the original text
tfidf = TfidfVectorizer(max_features=1000)
tfidf.fit(pd.concat([train_data['edit_text'], test_data['edit_text']]))  # Fit on all available text
X_train_tfidf = tfidf.transform(train_data['edit_text']).toarray()
X_test_tfidf = tfidf.transform(test_data['edit_text']).toarray()

# Sentiment Analysis directly on the original text
train_data['polarity'] = train_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
train_data['subjectivity'] = train_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
test_data['polarity'] = test_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
test_data['subjectivity'] = test_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Combining features
X_train = np.hstack((X_train_tfidf, train_data[['polarity', 'subjectivity']].values))
X_test = np.hstack((X_test_tfidf, test_data[['polarity', 'subjectivity']].values))

# Extracting labels
y_train = train_data['is_sockpuppet']
y_test = test_data['is_sockpuppet']

# Model Training using SVM
model = SVC(random_state=42)
model.fit(X_train, y_train)

# Predictions and Evaluation
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Integrated Semantic (SVM) and Sentiment (Random Forest) Analysis for Sockpuppet Detection in Wikipedia

This section employs a dual-model approach for sockpuppet detection in Wikipedia, utilizing **Support Vector Machines (SVM) for semantic analysis** and **Random Forest for sentiment analysis**. By integrating both analytical dimensions, the methodology seeks to provide a comprehensive view of the textual data, capturing both the explicit meaning of the text (semantics) through SVM and the underlying emotions or attitudes (sentiment) through Random Forest. This hybrid approach enhances the accuracy of sockpuppet identification by leveraging both linguistic structure and sentiment cues.

```python
# Importing necessary tools and libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob
from sklearn.svm import SVC  # SVM for Semantic Analysis
from sklearn.ensemble import RandomForestClassifier  # RF for Sentiment Analysis
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the pre-split training and testing datasets
train_data = pd.read_csv('wikipedia_sockpuppet_dataset_TRAIN.csv')
test_data = pd.read_csv('wikipedia_sockpuppet_dataset_TEST.csv')

# Feature extraction with TF-IDF for semantic analysis
tfidf = TfidfVectorizer(max_features=1000)
tfidf.fit(pd.concat([train_data['edit_text'], test_data['edit_text']]))  # Fit on all available text
X_train_semantic = tfidf.transform(train_data['edit_text']).toarray()
X_test_semantic = tfidf.transform(test_data['edit_text']).toarray()

# Sentiment Analysis feature extraction
train_data['polarity'] = train_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
train_data['subjectivity'] = train_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
test_data['polarity'] = test_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
test_data['subjectivity'] = test_data['edit_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Preparing sentiment-based feature set for Random Forest
X_train_sentiment = train_data[['polarity', 'subjectivity']].values
X_test_sentiment = test_data[['polarity', 'subjectivity']].values

# Extracting labels
y_train = train_data['is_sockpuppet']
y_test = test_data['is_sockpuppet']

# Standardize sentiment features for better performance
scaler = StandardScaler()
X_train_sentiment = scaler.fit_transform(X_train_sentiment)
X_test_sentiment = scaler.transform(X_test_sentiment)

# Model Training
# Train SVM on Semantic Features
svm_model = SVC(probability=True, random_state=42)  # Enable probability outputs
svm_model.fit(X_train_semantic, y_train)

# Train Random Forest on Sentiment Features
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train_sentiment, y_train)

# Get probability predictions from both models
svm_proba = svm_model.predict_proba(X_test_semantic)[:, 1]
rf_proba = rf_model.predict_proba(X_test_sentiment)[:, 1]

# Combine predictions: Averaging probabilities (you can change to weighted average)
final_proba = (svm_proba + rf_proba) / 2  # Simple ensemble strategy

# Convert probabilities to final class labels (threshold = 0.5)
y_pred_combined = (final_proba >= 0.5).astype(int)

# Classification report for the combined model
print("\nCombined SVM (Semantic) + RF (Sentiment) Model:")
print(classification_report(y_test, y_pred_combined))
print("Accuracy:", accuracy_score(y_test, y_pred_combined))
```
